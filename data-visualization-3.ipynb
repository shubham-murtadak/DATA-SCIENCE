# 2.Create a histogram for each feature in the dataset to illustrate the feature distributions.

A histogram is a traditional visualization tool that counts the number of data that fall into discrete bins to illustrate the distribution of one or more variables. This function can add a smooth curve derived using a kernel density estimate to the statistic computed within each bin to estimate frequency, density, or probability mass.

```python
sns.histplot(data=df, x="sepal_length", kde=True, color='red')
```

```python
sns.histplot(data=df,x=df['sepal_width'],kde=True,color='gold')
```

```python
sns.histplot(data=df,x=df['petal_width'],kde=True,color='green')
```

```python
sns.histplot(data=df,x='petal_width',kde=True,color='black')
```

```python
sns.histplot(data=df,x=df['species'],kde=True)
```

# 3.Create a boxplot for each feature in the dataset.

Boxplot is a convenient way to visualize the distribution of data through their quartiles. Box plots usually have vertical lines extending from the boxes which are termed as whiskers. These whiskers indicate variability outside the upper and lower quartiles, hence Box Plots are also termed as box-and-whisker plot and box-and-whisker diagram. Any Outliers in the data are plotted as individual points.

```python
sns.boxplot(df['sepal_length'])
```

```python
sns.boxplot(df['sepal_width'])
```

```python
sns.boxplot(df['petal_length'])
```

```python
sns.boxplot(df['petal_width'])
```

```python
sns.boxplot(df['species'])
```

```python
sns.boxplot(x='sepal_length',y='species',data=df)
```

```python
sns.boxplot(x='petal_length',y='species',data=df)
```
