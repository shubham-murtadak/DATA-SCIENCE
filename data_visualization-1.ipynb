# Understanding the Dataset - Titanic
The Titanic dataset is a classic dataset in data science with around 891 passengers and features like passenger class, age, sex, and most importantly, survival information. It's used for exploring data analysis techniques, building models to predict survival, and understanding factors that affected survival rates.

# Who were the Titanic passengers?
Before getting to the main question (who survived), let's take a look at the dataset to get a sense for how the observations are distributed into the different levels of our factors of interest.

## How many men, women, and children are in our sample?
First let's count the number of males and females, ignoring age.

```python
# Visualize survival by different features
sns.set(style="whitegrid")

# Visualize survival by sex
sns.catplot(x="sex", hue="survived", kind="count", data=titanic)
```

```python

# Visualize survival by class
sns.catplot(x="pclass", hue="survived", kind="count", data=titanic)
```

```python
# Visualize survival by age
sns.violinplot(x="survived", y="age", data=titanic, inner=None)
sns.swarmplot(x="survived", y="age", data=titanic, color="k", alpha=0.5)
```

# Survival by Embarked Port:

```python
sns.catplot(x="embarked", hue="survived", kind="count", data=titanic)
```

# Survival by Number of Siblings/Spouses:

```python
sns.catplot(x="sibsp", hue="survived", kind="count", data=titanic)
```

# Survival by Number of Parents/Children:

```python
sns.catplot(x="parch", hue="survived", kind="count", data=titanic)
```

# Survival by Fare:

```python
sns.histplot(x="fare", hue="survived", data=titanic, bins=30, kde=True)
```

# Survival by Age and Sex (Kernel Density Estimation - KDE):

```python
sns.displot(x="age", hue="survived", col="sex", kind="kde", data=titanic)
```

# 2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.

```python
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Plot histogram of ticket prices
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Distribution of Ticket Prices')
plt.show()
```
